```python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import numpy as np
from sklearn import linear_model
```

```python
modelA = sm.OLS.from_formula('Salary~Age', data =df)
resultA = modelA.fit()
resultA.summary()
```

### Observaciones:
* Variable depentiente: Salary
* Variable independiente: Age
* R-squared: 0.380, no es muy confiable porque explica solo 38% de Salary basado de variable Age
* Df Model: 1
* Intercept ... > 0.05, no es estadísiticamente significativo
--

`Salary = 349.4458 + 48.0932*Age`

```python
print('Salart of a person whose Age is 33: \t', 349.4458 + 48.0932*33)
df.iloc[:2,:]
```

```python
print('Diferencia es :\t', 1500-1936.5214)
print('Diferencia es :\t', 1200-1936.5214)
resultA.resid
```

```python
plt.figure(figsize = (15,10))
sns.set(font_scale = 1.7)

reg_plot=sns.regplot(x= 'Age',
            y = 'Salary',
            data = df,
            marker = '+',
            color = 'green',
            scatter_kws = {'s': 200}
           )
reg_plot.set_title('Regression plot of Salary vs Age :) ') 
reg_plot.text(40,1500, 'Salary = 349.4458 + 48.0932*Age',
              color = 'green',
              weight = 'semibold',
              size = 'medium'
             )
```

```python
RESIA= resultA.resid
RESIA
```

```python
resi_plot = sns.scatterplot(x = resultA.fittedvalues,
                            y = RESIA)
```

## conclusión
Los resiudales tienen un comportamiento normal y quiere decir que son aceptables

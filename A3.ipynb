# pivot tables

```python
import pandas as pd
import numpy as np
```

```python
%ls
```

```python
df = pd.read_csv('antropometria-dataset.csv')
df.head()
```

```python
df.columns
```

```python
my_pivot = pd.pivot_table(df,
                          index = ['desc_ent'],
                          values = ['peso'],
                          aggfunc = 'median',
                          columns = ['sexo']
                         )
my_pivot
```

```python
df = pd.read_csv('adult.csv')
df.head()
```

```python
df.columns
```

```python
my_pivot = pd.pivot_table(df,
                          index = ['sex','marital-status'],
                          values = ['age'],
                          aggfunc = 'median',
                          columns = ['education']
                         )
my_pivot
```

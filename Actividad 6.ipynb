## Pregunta de Investigación #1: ¿Como ha cambiado el porcentaje de población con acceso a electricidad en Colombia?

```python
df.loc['Colombia', years].plot()
plt.xlabel('years',fontsize =10)
plt.xticks(fontsize = 10, rotation = 0)

plt.ylabel('% de Población con Acceso a Electricidad', fontsize = 10)
plt.yticks(fontsize =10, rotation= 0)

plt.title('Población de Colombia con Acceso a Electricidad', fontsize = 13)
plt.show()
```

### Conclusión

La población de Colombia ha tenido, en general, una tendencia positiva a lo largo de los años. Ha habido algunos años donde hubo un crecimiento más pronunciado que otros y otros en donde ha habido una disminución en el % de población. Sin embargo en los últimos años, su porcentaje esta muy cercano, si no es que logró llegar al 100%.

## Pregunta de Investigación #2: ¿Cuáles son los top 10 países con el mejor porcentaje de acceso a la electricidad?

```python
df_topten=df.sort_values(by='Total Average', 
                          ascending=False,
                          axis=0,
                         )
df_topten=df_topten.head(10)
df_topten=df_topten.loc[:,'Total Average']

fig=plt.figure()

df_topten.plot(kind='bar')

plt.xticks(fontsize=12,rotation=45)
plt.yticks(fontsize=14,rotation=0)

plt.ylim((95,105))

plt.title('Países con más % de acceso de electricidad', fontsize=15)

plt.xlabel('País', fontsize=16)
plt.ylabel('% de población con acceso', fontsize=16)   
```

### Conclusión 
Todos los países con el mejor porcentaje de población con acceso a electricidad llegan al 100%, lo que quiere decir que todos sus habitantes tienen electricidad. 
Esta gráfica está compuesta principalmente por países primer mundistas y la mayoría se encuentra en Europa.

## Pregunta de Investigación #3: ¿Cómo se compara el porcentaje de población con acceso a electricidad de un país primer mundista y uno tercermundista?

```python
dfN = df.loc[['Norway', 'Mexico'],years]
dfN = dfN.transpose()
dfN.plot(kind ='line')
```

### Conclusión
Dado que los países primermundistas tienen más accesos a recursos, tienen la posibilidad de proporcionar con más facilidad acceso a la electricidad a toda su población. Por el otro lado, el porcentaje de población con acceso a electricidad a traves de los años en un país tercermundista va cambiando y, en este caso, es positivo. 

Claramente, dependiendo del país esta gráfica puede variar porque cada uno tiene un registro de datos diferente.

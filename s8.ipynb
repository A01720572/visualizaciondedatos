```python
import pandas as pd
```

```python
df = pd.read_csv('weight-height.csv')
df.head()
```

```python
df.corr() #para conocer la correlación
```

```python
x = df['Height']
y = df['Weight']
#type(y) #para saber que tipo de datos son, en este caso "series"
x = x.to_frame()
y = y.to_frame()
```

```python
pip install statsmodels
```

```python
import statsmodels.api as sm
```

```python
pip install sklearn
```

```python
from sklearn import linear_model
```

```python
lm = linear_model.LinearRegression()
lm
```

```python
model_lm = lm.fit(y,x)
```

```python
print('La intercepción es \t:', model_lm.intercept_)
print('Los coeficientes son \t:', model_lm.coef_)
print('El valor de R^2 es \t:', model_lm.score(x,y))
```

```python
models_ols = sm.OLS.from_formula('x~y', data = df) #Height = b0 + b1*Weight
result = models_ols.fit()
result.summary()
```

```python
numeros = ([200,201,94])
numeros = pd.DataFrame(numeros)
numeros
```

```python
y = model_lm.predict(numeros)
y = pd.DataFrame(y)
y
```
